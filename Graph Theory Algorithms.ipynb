### Importing relevant packages


```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
A= {
    'A':['C','D'],
    'B':['D','E'],
    'C':['A','D'],
    'D':['A','C','B','E'],
    'E':['B','D','F'],
    'F':['D','E']
    
}


G=nx.Graph(A)
```

### Graph

### DFS

```python
A= {
    'A':['C','D'],
    'B':['D','E'],
    'C':['A','D'],
    'D':['A','C','B','E'],
    'E':['B','D','F'],
    'F':['D','E']
    
}


G=nx.Graph(A)
```

```python
T = nx.depth_first_search.dfs_tree(G, 'A')

plt.figure(figsize=(10,8))
nx.draw(T, node_size=1000,with_labels=True)
plt.show()
```

### Dijkstra

```python
A= {
    'A':['C','D'],
    'B':['D','E'],
    'C':['A','D'],
    'D':['A','C','B','E'],
    'E':['B','D','F'],
    'F':['D','E']
    
}

G=nx.Graph(A)
```

```python
G.add_weighted_edges_from([
    ('A', 'C', 4),
    ('A', 'D', 4),
    ('C','D', 2),
    ('D', 'B', 3),
    ('D','F',2),
    ('D', 'E', 4),
    ('F', 'E', 3),
    ('B','E',3)
],weight='distance')
```

```python
G.nodes
```

```python
for j in G.nodes:
    for i in G.nodes:
        plt.figure(figsize=(8,6))
        S = nx.shortest_path(G, j,i, weight='distance')
        distance=nx.shortest_path_length(G,j,i,weight='distance')
        noCor = ["blue" if n in S else "red" for n in G.nodes()]
        pos = nx.spring_layout(G)
        nx.draw_networkx_nodes(G, pos=pos, node_color=noCor)
        nx.draw_networkx_edges(G, pos=pos)
        

        nx.draw(G,pos=pos,node_color=noCor,with_labels=True,node_size=1000)
        plt.title(f'From {j} to {i}, distance={distance}')
        plt.show()
```

### Prim

```python
A= {
    'A':['C','D'],
    'B':['D','E'],
    'C':['A','D'],
    'D':['A','C','B','E'],
    'E':['B','D','F'],
    'F':['D','E']
    
}

G=nx.Graph(A)
```

```python
G.add_weighted_edges_from([
    ('A', 'C', 4),
    ('A', 'D', 4),
    ('C','D', 2),
    ('D', 'B', 3),
    ('D','F',2),
    ('D', 'E', 4),
    ('F', 'E', 3),
    ('B','E',3)
],weight='distance')
```

```python
T = nx.minimum_spanning_tree(G,algorithm="prim")

plt.figure()
nx.draw(T, with_labels=True,node_size=1000)
plt.show()
```
